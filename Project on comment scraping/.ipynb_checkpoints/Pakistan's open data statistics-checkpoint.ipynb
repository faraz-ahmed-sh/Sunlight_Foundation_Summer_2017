## Global Open Data Index Scores

```python
Countries = ['Pakistan', 'India', 'Bangladesh']
```

```python
Score_2015 = [30, 55, 0] #in percentage
Score_2014 = [45, 68, 34] 
Score_2013 = [0, 40, 34]
```

```python
def multiple_chart(bar1, bar2, bar3, bar4, width, title):

    import numpy as np
    import matplotlib.pyplot as plt

    N = 3

    ind = np.arange(N)  # the x locations for the groups
    width      # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, bar1, width, color='r')

    rects2 = ax.bar(ind + width, bar2, width, color='y')
    rects3 = ax.bar(ind + width*2, bar3, width, color='g')
    
    if bar4 != None:
        rects4 = ax.bar(ind + width*3, bar4, width, color='b')

    # add some text for labels, title and axes ticks
    ax.set_ylabel('Scores')
    ax.set_title(title)
    ax.set_xticks(ind + width / 2)
    ax.set_ylim(0, 100)
    ax.set_xticklabels(Countries)
    
    if bar4 == None: 
        ax.legend((rects1[0], rects2[0], rects3[0]), ('2015', '2014', '2013'))
    else:
        ax.legend((rects1[0], rects2[0], rects3[0], rects4[0]), ('2016', '2015', '2014', '2013'))

    def autolabel(rects):
        """
        Attach a text label above each bar displaying its height
        """
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.00*height,
                    '%d' % int(height),
                    ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)
    autolabel(rects3)
    ax.set_facecolor('silver')
    #ax.set_facecolor((1, 0, 0))
    plt.show()
    fig.savefig(title + '.png', dpi=200)
```

```python
multiple_chart(Score_2013, Score_2014, Score_2015, None, 0.26, 'Global Open Data Index Scores')
```

## Open Data Barometer Scores

```python
ODB_Score_2016 = [4, 11, 43] #in percentage
ODB_Score_2015 = [6, 34, 7] #in percentage
ODB_Score_2014 = [13, 12, 33]
ODB_Score_2013 = [10, 10, 33]
```
